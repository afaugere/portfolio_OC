# VI) PreFeaturing

```python
import time
from datetime import datetime
```

```python
# Suppression des colonnes inutiles
df_merged.drop(["customer_id", "seller_id", "shipping_limit_date", 
                "product_category_name", "customer_zip_code_prefix"], axis = 1, inplace = True)
```

```python
df_merged
```

```python
# Cr√©ation d'une colone volume colis
df_merged["Volume_Colis_cm3"] = df_merged["product_length_cm"] *  df_merged["product_width_cm"] * df_merged["product_height_cm"]
```

```python
# Suppression des colonnes initiales
df_merged.drop(["product_length_cm", "product_width_cm", "product_height_cm"], axis = 1, inplace = True)
```

```python
# Cr√©ation colonne ==> Derni√®re commane
df_merged["Time_Last_cmde_H"] = datetime.now() - df_merged["order_purchase_timestamp"]
# Conversion en secondes
df_merged["Time_Last_cmde_H"] = df_merged["Time_Last_cmde_H"].apply(lambda x : x.total_seconds())
# Conversion en heures
df_merged["Time_Last_cmde_H"] = df_merged["Time_Last_cmde_H"]/3600
```

```python
# Cr√©ation colonne ==> D√©lai approbation commande en minutes
df_merged["Time_Approb_Order_H"] = df_merged["order_approved_at"]-df_merged["order_purchase_timestamp"]
# Conversion en secondes
df_merged["Time_Approb_Order_H"] = df_merged["Time_Approb_Order_H"].apply(lambda x : x.total_seconds())
# Conversion en heures
df_merged["Time_Approb_Order_H"] = df_merged["Time_Approb_Order_H"] / 3600
```

```python
# Cr√©ation colonne ==> D√©lai commande au transporteur
df_merged["Time_Approb_Carrier_H"] = df_merged["order_delivered_carrier_date"]-df_merged["order_approved_at"]
# Conversion en secondes
df_merged["Time_Approb_Carrier_H"] = df_merged["Time_Approb_Carrier_H"].apply(lambda x : x.total_seconds())
# Conversion en heures
df_merged["Time_Approb_Carrier_H"] = df_merged["Time_Approb_Carrier_H"] / 3600
```

```python
# Cr√©ation colonne ==> D√©lai de livraison estim√©
df_merged["Time_Delivery_Estimation_H"] = df_merged["order_estimated_delivery_date"]-df_merged["order_purchase_timestamp"]
# Conversion en secondes
df_merged["Time_Delivery_Estimation_H"] = df_merged["Time_Delivery_Estimation_H"].apply(lambda x : x.total_seconds())
# Conversion en heures
df_merged["Time_Delivery_Estimation_H"] = df_merged["Time_Delivery_Estimation_H"] / 3600
```

```python
# Cr√©ation colonne ==> D√©lai de livraison 
df_merged["Time_Delivery_H"] = df_merged["order_delivered_customer_date"]-df_merged["order_purchase_timestamp"]
# Conversion en secondes
df_merged["Time_Delivery_H"] = df_merged["Time_Delivery_H"].apply(lambda x : x.total_seconds())
# Conversion en heures
df_merged["Time_Delivery_H"] = df_merged["Time_Delivery_H"] / 3600
```

```python
# Cr√©ation colonne ==> D√©lai de mise commentaires Client √† la commande
df_merged["Time_Order_Review_H"] = df_merged["review_creation_date"]-df_merged["order_purchase_timestamp"]
# Conversion en secondes
df_merged["Time_Order_Review_H"] = df_merged["Time_Order_Review_H"].apply(lambda x : x.total_seconds())
# Conversion en heures
df_merged["Time_Order_Review_H"] = df_merged["Time_Order_Review_H"] / 3600
```

```python
# Cr√©ation colonne ==> D√©lai de mise commentaires Client √† livraison
df_merged["Time_Delivery_Review_H"] = df_merged["review_creation_date"]-df_merged["order_delivered_customer_date"]
# Conversion en secondes
df_merged["Time_Delivery_Review_H"] = df_merged["Time_Delivery_Review_H"].apply(lambda x : x.total_seconds())
# Conversion en heures
df_merged["Time_Delivery_Review_H"] = df_merged["Time_Delivery_Review_H"] / 3600
```

```python
# Cr√©ation colonne ==> D√©lai de r√©ponse
df_merged["Time_Answer_Review_H"] = df_merged["review_answer_timestamp"]-df_merged["review_creation_date"]
# Conversion en secondes
df_merged["Time_Answer_Review_H"] = df_merged["Time_Answer_Review_H"].apply(lambda x : x.total_seconds())
# Conversion en heures
df_merged["Time_Answer_Review_H"] = df_merged["Time_Answer_Review_H"] / 3600
```

```python
# Respect du d√©lai
df_merged["Delay_H"] = df_merged["Time_Delivery_Estimation_H"] - df_merged["Time_Delivery_H"]
```

```python
# densit√© colis
df_merged["Densit√©_colis_(g/cm3)"] = df_merged['product_weight_g']/df_merged['Volume_Colis_cm3']
```

```python
df_merged
```

```python
# Colonnes cat√©gorielles

def graph_count_categ_2(df,cols_categ) :
    i =0
    plt.rcParams["figure.figsize"] = [12,80]
    for col in cols_categ :
        i+=1
    # size
        if len(df[col].value_counts())<=20 :
            plt.subplot(10,1,i)
            # barplot 
            plt.bar(df[col].value_counts().index, df[col].value_counts().values, color = "blue")
            # Title and labels
            plt.title(f"values_counts_{col}")
            plt.xlabel("categories")
            plt.xticks(rotation = 90)
        else :
            top_values_counts = df_merged[col].value_counts().sort_values(ascending = False)[0:30]
            plt.subplot(10,1,i)
            # barplot 
            plt.bar(top_values_counts.index, top_values_counts.values, color = "blue")
            # Title and labels
            plt.title(f"top_20_values_counts_{col}")
            plt.xlabel("categories")
            plt.xticks(rotation = 90)
    plt.tight_layout()
    plt.show()
```

```python
# Colonnes num√©riques

def graph_box_plot_num_uni(df,cols_num) :
    i = 0
    plt.rcParams["figure.figsize"] = [12,80]
    for col in cols_num :
        i+=1
        # size
        plt.subplot(len(cols_num),1,i)
        # boxplot 
        sns.boxplot(data = df[col], orient = "h", showmeans = True)
        # Title and labels
        plt.title(f"box_plot_{col}")
        plt.xticks(rotation = 90)
    plt.tight_layout()
    plt.show()
```
