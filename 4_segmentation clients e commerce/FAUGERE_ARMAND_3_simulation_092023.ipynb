```python
# Importantation data et transformation
def importation_data():
    # importation
    df = pd.read_csv("data_origin.csv")
    # rajou d'une colonne pour filtrer sur les jours
    df["Time_last_cmde_J"] = df["Time_Last_cmde_H"]/24
    return(df)
```

# II) Featuring

```python
# Fontion de tri temporel pour rentrer dans l'algorithme
def tri_temporel(df,nb_jours):
    # dernière cmde de 2017 en heure 
    time_last_order_2017_j = 2122
    # filtre
    df = df.sort_values("Time_last_cmde_J")
    df = df[df["Time_last_cmde_J"]<=(time_last_order_2017_j + nb_jours)]
    return(df)
```

```python
# Variable temporelle en jour pour aller jusqu'à T
list_temporelle = np.arange(0,250,30)
```

```python
list_temporelle
```

## II.1) Aggrégation

```python
# fonction aggrégation Clients (calcul des variables sur l'intervale de temps donnnée)
def aggregation_client(df):
    # NB Total de cmde par Client
    df_count_cmde = df.groupby(["customer_unique_id"]).agg(["count"])["order_id"]
    # Sommes
    df_sum = df.groupby(["customer_unique_id"])[['payment_value','review_score_1','review_score_2',
                                                 'review_score_3', 'review_score_4', 'review_score_5',
                                                 'payment_type_boleto', 'payment_type_credit_card',
                                                 'payment_type_debit_card', 'payment_type_voucher']].sum()
    # Mean
    df_mean = df.groupby(["customer_unique_id"])[['Time_Answer_Review_H', 
                                                     'Delay_H', 'product_weight_g','Volume_Colis_cm3']].mean()
    # Min date cmde
    df_min = df.groupby(["customer_unique_id"]).agg(["min"])["Time_Last_cmde_H"]
    # Fusion count et somme
    df = df_count_cmde.merge(df_sum, how = "inner", on = "customer_unique_id")
    # Fusion df et min
    df = df.merge(df_min, how = "inner", on = "customer_unique_id")
    # Fusion df et mean
    df = df.merge(df_mean, how = "inner", on = "customer_unique_id")
    # renomage des colonnes
    df = df.rename(columns = {'count' : 'Nb_cmde', 'min' : 'Time_last_cmde_H' , 'payment_value' : 'payment_value_sum',
                     'review_score_1' : 'review_score_1_sum', 'review_score_2' : 'review_score_2_sum',
                     'review_score_3' : 'review_score_3_sum','review_score_4' : 'review_score_4_sum',
                     'review_score_5' : 'review_score_5_sum', 'payment_type_boleto' : 'payment_type_boleto_sum',
                     'payment_type_credit_card' : 'payment_type_credit_card_sum',
                     'payment_type_debit_card' : 'payment_type_debit_card_sum', 
                     'payment_type_voucher' : 'payment_type_voucher_sum','Delay_H' : 'Delay_H_mean',
                     'product_weight_g' : 'product_weight_g_mean', 'Volume_Colis_cm3' : 'Volume_Colis_cm3_mean'})
    return(df)
```

## II.2) Normalisation

```python
# importation scaler
from sklearn.preprocessing import StandardScaler
```

```python
# Normalisation 
def standardisation(df) :
    # assignation du scaler
    scaler = StandardScaler()
    # dataframe de sortie
    df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
    return(df) 
```

## II.3) KMEANS

```python
# importation KMeans
from sklearn.cluster import KMeans
# importation ARI
from sklearn.metrics.cluster import adjusted_rand_score
```

```python
# fonction ARI 
def ARI_graph(ARI_list, list_temporelle, nom_cluster):   
# Graphique plot
    plt.rcParams['figure.figsize']=[12,6]
    plt.plot(list_temporelle, ARI_list, color = "b")
    plt.title(f"ARI clusters {nom_cluster}")
    plt.xlabel("temps")
    plt.tight_layout()
    plt.show()
```

### II.3.1) Sans PCA

##### 'Nb_cmde', payment_value_sum'

```python
# Focntion de clustering cluster_Nb_cmde_payment_value_sum
def cluster_Nb_cmde_payment_value_sum(df) :
    # SélEction des colonnes pertinentes (Fréquence, Montant)
    X = df[['Nb_cmde', 'payment_value_sum']]
    # 3 clusters retenus, mise en place kmeans
    kmeans_Ncmde_payS = KMeans(n_clusters = 3, random_state=0, n_init = 10)
    # clustering
    kmeans_Ncmde_payS.fit_predict(X)
    return kmeans_Ncmde_payS
```

```python
# Entrainement M0
# fonction d'importation des datas
df = importation_data()
# fonction tri temporel au dataframe
df = tri_temporel(df,0)
# fonction aggrégation Clients (calcul des variables sur l'intervale de temps donnnée)
df = aggregation_client(df)
# fonction standardisation
df = standardisation(df)
# fonction de clustering
M0 = cluster_Nb_cmde_payment_value_sum(df)
```

```python
ARI_Nb_cmde_payment_value_sum = []
# Boucle temporelle et de réalisation de fonctions
for nb_jours in list_temporelle :
    # fonction d'importation des datas
    df = importation_data()
    # fonction tri temporel au dataframe
    df = tri_temporel(df,nb_jours)
    # fonction aggrégation Clients (calcul des variables sur l'intervale de temps donnnée)
    df = aggregation_client(df)
    # fonction standardisation
    df = standardisation(df)
    # fonction de clustering pour I
    MI = cluster_Nb_cmde_payment_value_sum(df)
    # préductions avec modèle M0
    label_M0 = M0.predict(df[['Nb_cmde', 'payment_value_sum']])
    # remplissage liste
    ARI_Nb_cmde_payment_value_sum.append(adjusted_rand_score(MI.labels_,label_M0))
```

```python
# liste ARI
ARI_Nb_cmde_payment_value_sum
```

```python
# Graphique ARI
ARI_graph(ARI_Nb_cmde_payment_value_sum,list_temporelle,"Nb_cmde_payment_value_sum")
```

### II.3.2) Avec PCA

```python
from sklearn.decomposition import PCA
```

##### Factorisation focus économie

```python
# Focntion de pca_economique
def pca_economique(df) :
# PCA
    # Assignation des variables
    Xpca = df[['Nb_cmde', 'payment_value_sum','Time_last_cmde_H']]
    # Sélèction des 2 dimensions
    pca = PCA(2)
    pca.fit(Xpca)
    # Nouveaux vecteurs F1 et F2
    pca.transform(Xpca)
    # création df F1 et F2
    df_cluster_eco = pd.DataFrame({"F1" : pca.transform(Xpca)[:,0],"F2" : pca.transform(Xpca)[:,1]})
    return df_cluster_eco

# Focntion de clustering pca_economique
def cluster_pca_economique(df) :
    # KMEANS    
    # SélEction des colonnes pertinentes (payment_value_sum, Récense)
    X = df[['F1', 'F2']]
    # 4 clusters retenus
    kmeans_pca_eco = KMeans(n_clusters = 4, random_state=0, n_init = 10)
    # clustering
    kmeans_pca_eco.fit_predict(X)
    return kmeans_pca_eco
```

```python
# Entrainement M0
# fonction d'importation des datas
df = importation_data()
# fonction tri temporel au dataframe
df = tri_temporel(df,0)
# fonction aggrégation Clients (calcul des variables sur l'intervale de temps donnnée)
df = aggregation_client(df)
# fonction standardisation
df = standardisation(df)
# fonction pca
df = pca_economique(df)
# fonction de clustering   
M0 = cluster_pca_economique(df)
```

```python
ARI_pca_economique = []
# Boucle temporelle et de réalisation de fonctions
for nb_jours in list_temporelle :
    # fonction d'importation des datas
    df = importation_data()
    # fonction tri temporel au dataframe
    df = tri_temporel(df,nb_jours)
    # fonction aggrégation Clients (calcul des variables sur l'intervale de temps donnnée)
    df = aggregation_client(df)
    # fonction standardisation
    df = standardisation(df)
    # fonction pca
    df = pca_economique(df)
    # fonction de clustering pour I
    MI = cluster_pca_economique(df)
   # prédictions avec modèle M0
    label_M0 = M0.predict(df[['F1', 'F2']])
    # remplissage liste
    ARI_pca_economique.append(adjusted_rand_score(MI.labels_,label_M0))
```

```python
# Liste ARI
ARI_pca_economique
```

```python
# Graphique ARI
ARI_graph(ARI_pca_economique,list_temporelle,"pca_economique")
```

##### Factorisation satisfaction

```python

# Focntion  pca_economique
def pca_satisfaction(df) :
# PCA
    # Assignation des variables
    Xpca = df[['review_score_1_sum','review_score_5_sum','Time_last_cmde_H', 'Time_Answer_Review_H']]
    # Sélèction des 2 dimensions
    pca = PCA(2)
    pca.fit(Xpca)
    # Nouveaux vecteurs F1 et F2
    pca.transform(Xpca)
    # création df F1 et F2
    df_cluster_satisfaction = pd.DataFrame({"F1" : pca.transform(Xpca)[:,0],"F2" : pca.transform(Xpca)[:,1]})
    return df_cluster_satisfaction

# Focntion de clustering pca_economique
def cluster_pca_satisfaction(df) :    
    # KMEANS    
    # SélEction des colonnes pertinentes (payment_value_sum, Récense)
    X = df[['F1', 'F2']]
    # 3 clusters retenus
    kmeans_pca_satisfaction = KMeans(n_clusters = 3, random_state=0, n_init = 10)
    # clustering
    kmeans_pca_satisfaction.fit_predict(X)
    return kmeans_pca_satisfaction  
```

```python
# Entrainement M0
# fonction d'importation des datas
df = importation_data()
# fonction tri temporel au dataframe
df = tri_temporel(df,0)
# fonction aggrégation Clients (calcul des variables sur l'intervale de temps donnnée)
df = aggregation_client(df)
# fonction standardisation
df = standardisation(df)
# fonction pca
df = pca_satisfaction(df)
# fonction de clustering   
M0 = cluster_pca_satisfaction(df)
```

```python
ARI_pca_satisfaction = []
# Boucle temporelle et de réalisation de fonctions
for nb_jours in list_temporelle :
    # fonction d'importation des datas
    df = importation_data()
    # fonction tri temporel au dataframe
    df = tri_temporel(df,nb_jours)
    # fonction aggrégation Clients (calcul des variables sur l'intervale de temps donnnée)
    df = aggregation_client(df)
    # fonction standardisation
    df = standardisation(df)
    # fonction pca
    df = pca_satisfaction(df)
    # fonction de clustering pour I
    MI = cluster_pca_satisfaction(df)
    # prédictions avec modèle M0
    label_M0 = M0.predict(df[['F1', 'F2']])
    # remplissage liste
    ARI_pca_satisfaction.append(adjusted_rand_score(MI.labels_,label_M0))
```
